```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', 200)
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 300)
```

```python
mexico = pd.read_csv("mexico_joined.csv")
argentina = pd.read_csv("argentina_joined.csv")
venezuela = pd.read_csv("venezuela_joined.csv")
chile = pd.read_csv("chile_joined.csv")
uruguay = pd.read_csv("uruguay_joined.csv")
peru = pd.read_csv("peru_joined.csv")
colombia = pd.read_csv("colombia_joined.csv")
```

```python
mexico['Country'] = 'Mexico'
argentina['Country'] = 'Argentina'
venezuela['Country'] = 'Venezuela'
chile['Country'] = 'Chile'
uruguay['Country'] = 'Uruguay'
peru['Country'] = 'Peru'
colombia['Country'] = 'Colombia'
```

```python
frames = [mexico, argentina, venezuela, chile, uruguay, peru, colombia]

result = pd.concat(frames)
```

```python
import altair as alt
alt.Chart(result).mark_area().encode(
    alt.X('index:T'),
    y = alt.Y("day_date:Q", stack=None),
    color = "Country",
    tooltip = ('index:T', 'day_date', 'Country')
).properties(
    width=850,
    height=300
)
```

```python
import altair as alt
alt.Chart(result).mark_area(opacity=0.3).encode(
    x="index:T",
    y=alt.Y("day_date:Q", stack=None),
    color="Country",
    tooltip = ('index:T', 'day_date', 'Country')
).properties(
    width=850,
    height=300
)
```
